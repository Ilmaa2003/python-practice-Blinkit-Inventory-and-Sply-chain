```python
import numpy as np # scientific and mathematical calculations
import pandas as pd # data manipulation
import seaborn as sns # visualization
import matplotlib.pylab as plt # visualization
```

```python
orders = pd.read_csv("Orders.csv") # read file within pandas data frame
orders
```

```python
orders.info()
```

```python
#fixing data types

#if not data type
#suppose quantity was a string
#orders ["Quantity"] = orders["Quantity"].astype("int")

#if date time
orders["OrderDateTime"] = pd.to_datetime(orders["OrderDateTime"])

```

```python
orders.info()
```

```python
orders["ExpectedDeliveryTime"] = pd.to_datetime(orders["ExpectedDeliveryTime"])
orders["ActualDeliveryTime"] = pd.to_datetime(orders["ActualDeliveryTime"])
```

```python
orders.info()
```

```python
#to substract
orders["Actual_Delivery_inMinutes"] = (orders["ActualDeliveryTime"]- orders["OrderDateTime"]).dt.total_seconds()/60
orders
```

```python

orders["Expected_Delivery_inMinutes"] = (orders["ExpectedDeliveryTime"]- orders["OrderDateTime"]).dt.total_seconds()/60
orders
```

```python
#unique for distinct values
orders["Expected_Delivery_inMinutes"].unique()
```

```python
#unique for distinct values
orders["StoreID"].unique()
```

```python
orders["difference_between_expected_and_actual"] = orders["Expected_Delivery_inMinutes"] - orders["Actual_Delivery_inMinutes"]
orders
```

```python

#loc allows to work with indexes


orders.loc[(orders["difference_between_expected_and_actual"] == 0), "order_status"]="on time"
orders.loc[(orders["difference_between_expected_and_actual"] < 0), "order_status"]="Delayed"

orders
```

```python
sns.countplot(x="order_status", data=orders)
plt.show()
```

```python
plt.figure(figsize=(16,4))
sns.countplot(x="StoreID", hue="order_status", data=orders) # hue allows to provide legend
plt.xticks(rotation=90) # to rotate x labels to 90 degree
plt.show() # to visualize
```
